## 1. csv 파일로 만들기

```python
with open("bank-additional-full.csv","r") as fp:
    raw = fp.read().replace('"','')
    
    raw_split = []
    lines = raw.split("\n")
    title = lines[0].split(";")
    
    for line in lines[1:]:
        elements = line.split(";")
        ele_dict = {}
        
        for i in range(len(elements)):
            ele_dict[title[i]] = elements[i]
        
        raw_split.append(ele_dict)
```

```python
raw_split[:2]
```

```python
len(raw_split)
# 맨 마지막 행이 empty로 들어가 있어서 카운팅이 +1 됨. 최종적으로 마지막 행 지우기
```

```python
raw_split = raw_split[:-1]
```

```python
with open("bank-info-preprocessing.csv", 'w') as fw:
    for t in title:
        fw.write(t+",")
    
    fw.write("\n")
    
    for e in raw_split:
        for t in title:
            fw.write(e[t]+",")
        fw.write("\n")
```

```python
x_train = train_set.values[:,0:-1]
y_train = train_set.values[:,[-1]]
```

```python
x_train_l = x_train.tolist()
```

```python
x_train_l
```

```python
for t in x_train_l:
    for i in [1, 2, 3, 4, 5, 6, 7, 8, 9, 12]:
        t[i] = t[i].tolist()
```

```python
x_train_l
```

```python
x_train_flat = []
for l in x_train_l:
    d = []
    for n in range(len(l)):
        if type(l[n]) != list :
            d.append(l[n])
        else :
            for k in range(len(l[n])):
                d.append(l[n][k])
    x_train_flat.append(d)
```

```python
x_train = np.array(x_train_flat)
```

```python
x_train.shape
```

```python
y_train_l = y_train.tolist()
```

```python
for t in y_train_l:
    t[0] = t[0].tolist()
```

```python
y_train_l = np.array(y_train_l)
```

```python
y_train=y_train_l.reshape(-1,1)
```

```python
y_train
```

```python
y_train.shape
```

```python
X = tf.placeholder(tf.float32, shape=[None,60])
Y = tf.placeholder(tf.float32, shape=[None,1])
```

```python
W = tf.Variable(tf.random_normal([60,1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
hypothesis = tf.sigmoid(tf.matmul(X,W)+b)
cost = -tf.reduce_mean(Y*tf.log(hypothesis)+(1-Y)*tf.log(1-hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        cost_val, _ = sess.run([cost, train], feed_dict={X:x_train.astype(np.float32), Y:y_train.astype(np.float32)})
        if step % 200 == 0:
            print(step, cost_val)
            
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={X:x_train, Y:y_train})
print("\nHypothesis: ", h, "\nCorrect: ", c, "\nAccuracy: ", a)
```
